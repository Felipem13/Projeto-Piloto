### Passo 1 - Importar Arquivos e Bibliotecas

```python
import pandas as pd 
from pathlib import Path
import win32com.client as win32
```

```python
#importando tabelas
tabela_email = pd.read_excel(r'Bases de Dados/Emails.xlsx')
tabela_vendas = pd.read_excel(r'Bases de Dados/Vendas.xlsx')
tabela_lojas = pd.read_csv(r'Bases de Dados/Lojas.csv',sep=';',encoding='latin-1')

# ajustando tipo de coluna tabela lojas[ID loja] para int64
tabela_lojas['ID Loja'] = tabela_lojas['ID Loja'].astype('int64')
#tabela_lojas['ID Loja'] = tabela_lojas['ID Loja'].astype('int64')
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# inserir nome das lojas em tabela_vendas (mesclar)
tabela_vendas = tabela_vendas.merge(tabela_lojas,on='ID Loja')
```

```python
# dicionário vazio para armazenar tabelas
dic_lojas = {}
#percorrer coluna lojas em tabela de lojas
for loja in tabela_lojas['Loja']:
    # filtrar onde o nome loja é igual na coluna Loja (tabela_vendas), adicionar no dicionário
    # cada dic_lojas é um DataFrame
    dic_lojas[loja] = tabela_vendas.loc[tabela_vendas['Loja']==loja , :]
```

```python
# identificar qual a data mais recente
dia_indicador = tabela_vendas['Data'].max()
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
caminho_backup = r'Backup Arquivos Lojas'

for loja in dic_lojas:
    Path(f'{caminho_backup}/{loja}').mkdir(exist_ok=True,parents=True)
    
    nome_arquivo = f'{dia_indicador.day}_{dia_indicador.month}_{loja}.xlsx'
    dic_lojas[loja].to_excel(f'{caminho_backup}/{loja}/{nome_arquivo}',index=False)
    
# lista_nomes_arq = [arquivo.name for arquivo in Path('Backup Arquivos Lojas').iterdir()]
```

### Passo 4 - Calcular o indicador para 1 loja

```python
loja = 'Norte Shopping'
vendas_lojas_ano = dic_lojas[loja]
vendas_lojas_dia = vendas_lojas_ano.loc[vendas_lojas_ano['Data']==dia_indicador,:]

#faturamento
faturamento_ano = vendas_lojas_ano['Valor Final'].sum()
faturamento_dia = vendas_lojas_dia['Valor Final'].sum()

# Diversidade de Produtos 
qntd_produtos_ano = len(vendas_lojas_ano['Produto'].unique())
qntd_produtos_dia = len(vendas_lojas_dia['Produto'].unique())

#Ticket Médio Ano
valor_por_venda = vendas_lojas_ano[['Código Venda','Valor Final']].groupby('Código Venda').sum()
tickt_médio_ano = valor_por_venda['Valor Final'].mean()

valor_por_venda = vendas_lojas_dia[['Código Venda','Valor Final']].groupby('Código Venda').sum()
tickt_médio_dia = valor_por_venda['Valor Final'].mean()
```
